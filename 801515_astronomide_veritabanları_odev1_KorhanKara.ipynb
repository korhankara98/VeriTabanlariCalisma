# Soru 3 (10 puan):

Kataloğu sadece size verilen gezegen türü için sınırlandırınız.

```python
import numpy as np
import pandas as pd

Mdun = 5.972e24 #kg
Mjup = 1.898e27 #kg

Rdun = 6371 #km
Rjup = 69911 #km

Mdj = Mdun/Mjup  # Dünya'nın Jüpiter kütlesi cinsinden kütlesi
Rdj = Rdun/Rjup  # Dünya'nın Jüpiter yarıçapı cisinden yarıçapı

MDun_alt_sinir = 2 * Mdj
MDun_ust_sinir = 5 * Mdj

RDun_alt_sinir = 0.8 * Rdj
RDun_ust_sinir = 1.25 * Rdj

path = r'C:\Users\korhan\Desktop\Yüksek Lisans\VeriTabanları\Ödev 1\exo.csv' ###########
data = pd.read_csv(path)

kutleler = data.iloc[:, 2]
yaricaplar = data.iloc[:, 8]

kutle_kriteri = (kutleler > MDun_alt_sinir) & (kutleler < MDun_ust_sinir)
yaricap_kriteri = (yaricaplar > RDun_alt_sinir) & (yaricaplar < RDun_ust_sinir)

uygun_gezegenler = kutle_kriteri | yaricap_kriteri

print(data[uygun_gezegenler])
```

# Soru 4 (15 puan):

Söz konusu gezegen türü için kütle ve yarıçapı belirlenmiş olanların yoğunluklarını hesaplayarak, barınak yıldızın metal bolluğuna karşı çizdiriniz. Grafiğinizdeki aykırı noktaları belirleyerek çıkarınız. Grafiği çizdirmeden önce nasıl bir ilişki görmeyi beklerdiniz? Beklentiniz ile çizdirdiğiniz grafiği karşılaştırarak yorumlayınız. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Mdun = 5.972e24  # kg
Mjup = 1.898e27  # kg

Rdun = 6371  # km
Rjup = 69911  # km

Mdj = Mdun / Mjup  # Dünya'nın Jüpiter kütlesi cinsinden kütlesi
Rdj = Rdun / Rjup  # Dünya'nın Jüpiter yarıçapı cisinden yarıçapı

MDun_alt_sinir = 2 * Mdj
MDun_ust_sinir = 5 * Mdj

RDun_alt_sinir = 0.8 * Rdj
RDun_ust_sinir = 1.25 * Rdj

path = r'C:\Users\korhan\Desktop\Yüksek Lisans\VeriTabanları\Ödev 1\exo.csv'
data = pd.read_csv(path)

kutleler = data.iloc[:, 2]
yaricaplar = data.iloc[:, 8]
kutleler2 = (kutleler > MDun_alt_sinir) & (kutleler < MDun_ust_sinir)
yaricaplar2 = (yaricaplar > RDun_alt_sinir) & (yaricaplar < RDun_ust_sinir)
uygun_veri = kutleler2 | yaricaplar2

uygun_data = data[uygun_veri]  

metaller = uygun_data.iloc[:, 79]  
kutleler = uygun_data.iloc[:, 2]
yaricaplar = uygun_data.iloc[:, 8]
print(len(metaller))
print(len(kutleler))

yoğunluklar = kutleler / (4 / 3 * np.pi * np.power(yaricaplar, 3))
print(len(yoğunluklar))

nan_mask = np.isnan(metaller) | np.isnan(yoğunluklar)
metaller = metaller[~nan_mask]
yoğunluklar = yoğunluklar[~nan_mask]

Q1 = np.percentile(yoğunluklar, 5)
Q3 = np.percentile(yoğunluklar, 95)
IQR = Q3 - Q1
alt_sinir = Q1 - 1.5 * IQR
üst_sinir = Q3 + 1.5 * IQR
aykırılar = (yoğunluklar < alt_sinir) | (yoğunluklar > üst_sinir)

plt.figure(figsize=(10, 6))
plt.scatter(metaller, yoğunluklar, label='Uygun Gezegenler', color='blue')
plt.scatter(metaller[aykırılar], yoğunluklar[aykırılar], label='Aykırı Noktalar', color='red')
plt.xlabel('Metal Bolluğu')
plt.ylabel('Yoğunluk')
plt.title('Metal Bolluğuna Karşı Yoğunluk')
plt.legend()
plt.grid(True)
plt.show()
```

# Soru 7 (20 puan):

Kutu grafiği temelinde aykırı noktalarından ayıkladığınız parametrelerin histogramlarını çizdiriniz.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Mdun = 5.972e24  # kg
Mjup = 1.898e27  # kg

Rdun = 6371  # km
Rjup = 69911  # km

Mdj = Mdun / Mjup  # Dünya'nın Jüpiter kütlesi cinsinden kütlesi
Rdj = Rdun / Rjup  # Dünya'nın Jüpiter yarıçapı cisinden yarıçapı

MDun_alt_sinir = 2 * Mdj
MDun_ust_sinir = 5 * Mdj

RDun_alt_sinir = 0.8 * Rdj
RDun_ust_sinir = 1.25 * Rdj

path = r'C:\Users\korhan\Desktop\Yüksek Lisans\VeriTabanları\Ödev 1\exo.csv'
data = pd.read_csv(path)

kutleler = data.iloc[:, 2]
yaricaplar = data.iloc[:, 8]
kutleler2 = (kutleler > MDun_alt_sinir) & (kutleler < MDun_ust_sinir)
yaricaplar2 = (yaricaplar > RDun_alt_sinir) & (yaricaplar < RDun_ust_sinir)
uygun_veri = kutleler2 | yaricaplar2

uygun_data = data[uygun_veri]  

teff = uygun_data.iloc[:, 92]
metaller = uygun_data.iloc[:, 79]
Star_radii = uygun_data.iloc[:, 85]
Star_age = uygun_data.iloc[:, 89]

nan_mask = np.isnan(metaller) | np.isnan(teff) | np.isnan(Star_age) | np.isnan(Star_radii)
metaller = metaller[~nan_mask]
teff = teff[~nan_mask]
Star_age = Star_age[~nan_mask]
Star_radii = Star_radii[~nan_mask]

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(metaller, bins=20, edgecolor='black', alpha=0.7)
plt.title('Metallicity')
plt.xlabel('Metallicity')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(teff, bins=20, edgecolor='black', alpha=0.7)
plt.title('Effective Temperature (K)')
plt.xlabel('Effective Temperature (K)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(Star_age, bins=20, edgecolor='black', alpha=0.7)
plt.title('Star Age (Gyr)')
plt.xlabel('Star Age (Gyr)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(Star_radii, bins=20, edgecolor='black', alpha=0.7)
plt.title('Star Radius (Solar Radii)')
plt.xlabel('Star Radius (Solar Radii)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

def remove_outliers(data, threshold=1.5):
    q1 = np.percentile(data, 30)
    q3 = np.percentile(data, 70)
    iqr = q3 - q1
    lower_bound = q1 - threshold * iqr
    upper_bound = q3 + threshold * iqr
    return data[(data >= lower_bound) & (data <= upper_bound)]

metaller_filtered = remove_outliers(metaller)
teff_filtered = remove_outliers(teff)
Star_age_filtered = remove_outliers(Star_age)
Star_radii_filtered = remove_outliers(Star_radii)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(metaller_filtered, bins=20, edgecolor='black', alpha=0.7)
plt.title('Metallicity (Outliers Removed)')
plt.xlabel('Metallicity')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(teff_filtered, bins=20, edgecolor='black', alpha=0.7)
plt.title('Effective Temperature (K) (Outliers Removed)')
plt.xlabel('Effective Temperature (K)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(Star_age_filtered, bins=20, edgecolor='black', alpha=0.7)
plt.title('Star Age (Gyr) (Outliers Removed)')
plt.xlabel('Star Age (Gyr)')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(Star_radii_filtered, bins=20, edgecolor='black', alpha=0.7)
plt.title('Star Radius (Solar Radii) (Outliers Removed)')
plt.xlabel('Star Radius (Solar Radii)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

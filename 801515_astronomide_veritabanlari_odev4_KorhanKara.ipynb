Size verilen gruptaki gezegenlerin barınak yıldızlarının homojen bir veri setini oluşturmak üzere Gaia veritabanından yıldızlarınızın etkin sıcaklık değerlerini (<i>teff_gspphot</i>) çekiniz ve bir keman diyagramını çizdiriniz. Bu gruptaki gezegenler için barınak yıldız etkin sıcaklığının ortalama değeri ve standart sapmasını verdikten sonra göze çarpan bir trend varsa belirtiniz ve yorumlayınız. (15 puan)

3. HR diyagramında göze çarpan bir trend görüyor musunuz? Yorumlayınız. (20 puan)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astroquery.gaia import Gaia
from astroquery.simbad import Simbad
import astropy.units as u
from astropy.coordinates import SkyCoord
import numpy.ma as ma
import seaborn as sns
from astropy.constants import R_jup, M_jup
```

```python
file_path = 'C:\\Users\\korhan\\Desktop\\Ödev 3\\exonasa.csv'
df1 = pd.read_csv(file_path, header=0, comment='#')

radius_column = df1.columns[19]
mass_column = df1.columns[27]
discovery_method_column = 'discoverymethod'

filtered_df = df1[
    (df1[radius_column] >= 0.8) & (df1[radius_column] <= 1.25) &
    (df1[mass_column] >= 2) & (df1[mass_column] <= 5)
]


transit_df = filtered_df[filtered_df[discovery_method_column] == 'Transit']
radius_values = transit_df[radius_column].dropna()
mean_radius = radius_values.mean()
std_radius = radius_values.std()


plt.hist(radius_values, bins=8, edgecolor='black')
plt.axvline(mean_radius, color='r', linestyle='dashed', linewidth=1)
plt.title('Transit Yöntemiyle Keşfedilen Gezegenlerin Yarıçap Dağılımı')
plt.xlabel('Yarıçap (Jüpiter Yarıçapı)')
plt.ylabel('Gezegen Sayısı')
plt.show()

print(f"Ortalama Yarıçap: {mean_radius:.2f}")
print(f"Standart Sapma: {std_radius:.2f}")
```

```python
radius_values = filtered_df[radius_column].dropna()
mean_radius = radius_values.mean()
std_radius = radius_values.std()

ra_column = 'ra'
dec_column = 'dec'

ra = np.array(filtered_df[ra_column])
dec = np.array(filtered_df[dec_column])
width = u.Quantity(0.01, u.deg)
height = u.Quantity(0.01, u.deg)

teff = []

for i in range(len(ra)):
    coord = SkyCoord(ra=ra[i], dec=dec[i], unit=(u.degree, u.degree), frame='icrs')
    query = Gaia.query_object_async(coordinate=coord, width=width, height=height)
    if 'teff_gspphot' in query.colnames and not ma.is_masked(query['teff_gspphot'][0]):
        teff.append(query['teff_gspphot'][0])

star_temp_series = pd.Series(teff)
mean_star_temp = star_temp_series.mean()
std_star_temp = star_temp_series.std()

plt.figure(figsize=(10, 6))
sns.violinplot(y=star_temp_series, inner="quart")
plt.axhline(mean_star_temp, color='r', linestyle='dashed', linewidth=1)
plt.title('Keşfedilen Gezegenlerin Barınak Yıldızlarının Etkin Sıcaklık Dağılımı')
plt.ylabel('Etkin Sıcaklık (K)')
plt.text(0, mean_star_temp, f'Ortalama: {mean_star_temp:.2f} K', color='r', ha='left')
plt.show()

print(f"Ortalama Etkin Sıcaklık: {mean_star_temp:.2f} K")
print(f"Standart Sapma: {std_star_temp:.2f} K")
```

```python
star_radius_column = 'st_rad'
star_radius = np.array(filtered_df[star_radius_column])
teff = []
parallax = []
mag = []

for i in range(len(ra)):
    coord = SkyCoord(ra=ra[i], dec=dec[i], unit=(u.degree, u.degree), frame='icrs')
    query = Gaia.query_object_async(coordinate=coord, width=width, height=height)
    
    if not (ma.is_masked(query['parallax'][0]) or ma.is_masked(query['phot_g_mean_mag'][0]) or ma.is_masked(query['teff_gspphot'][0])):
        parallax.append(query['parallax'][0])
        mag.append(query['phot_g_mean_mag'][0])
        teff.append(query['teff_gspphot'][0])

def abs_mag(par, m):
    mv = m - (2.5 * np.log10((par / 10)**2))
    return mv

abs_magn = [abs_mag(parallax[k], mag[k]) for k in range(len(parallax))]
marker_sizes = star_radius * 100

plt.figure(figsize=(11, 8))
sc = plt.scatter(teff, abs_magn, c=abs_magn, cmap='viridis', edgecolor='k', s=marker_sizes, alpha=0.7)
cbar = plt.colorbar(sc)
cbar.set_label('Mutlak Parlaklık')
cbar.ax.invert_yaxis() 
plt.xlabel('Etkin Sıcaklık (K)', fontsize=14)
plt.ylabel('Mutlak Parlaklık', fontsize=14)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title('Süper Dünyaların Barınak Yıldızlarına Ait HR Diagramı', fontsize=16)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
jupiter_mass = M_jup.to('g').value
jupiter_radius = R_jup.to('cm').value

radius_column = 'pl_radj'
mass_column = 'pl_bmassj'
discovery_method_column = 'discoverymethod'
orbital_semi_major_axis_column = 'pl_orbsmax'
star_radius_column = 'st_rad'

planet_mass_g = filtered_df[mass_column] * jupiter_mass
planet_radius_cm = filtered_df[radius_column] * jupiter_radius
planet_volume_cm3 = (4/3) * np.pi * (planet_radius_cm**3)
planet_density_g_cm3 = planet_mass_g / planet_volume_cm3

semi_major_axis_au = filtered_df[orbital_semi_major_axis_column]
star_radius_solar = filtered_df[star_radius_column]
a_R = semi_major_axis_au / star_radius_solar

og = filtered_df.copy()
og['pl_density'] = planet_density_g_cm3
og['a_R'] = a_R

plt.figure(figsize=(10, 6))
plt.scatter(og['a_R'], og['pl_density'], c='b', alpha=0.6, edgecolors='k')
plt.xlabel('a / R*', fontsize=14)
plt.ylabel('Average Planet Density (g/cm³)', fontsize=14)
plt.title('a/R* vs Average Planet Density', fontsize=16)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
orbital_period_column = 'pl_orbper'  # Yörünge dönemi
eccentricity_column = 'pl_orbeccen'  # Yörünge dışmerkezliliği
mass_column = 'pl_bmassj'            # Kütle


plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
plt.boxplot(filtered_df[orbital_period_column])
plt.xlabel('Orbital Period (days)')
plt.title('Orbital Period Distribution')

plt.subplot(1, 3, 2)
plt.boxplot(filtered_df[eccentricity_column])
plt.xlabel('Eccentricity')
plt.title('Orbital Eccentricity Distribution')

plt.subplot(1, 3, 3)
plt.boxplot(filtered_df[mass_column])
plt.xlabel('Mass (Jupiter Masses)')
plt.title('Planetary Mass Distribution')

plt.tight_layout()
plt.show()
```

```python
discovery_counts = filtered_df[discovery_method_column].value_counts()

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
log_discovery_counts = np.log10(discovery_counts)
log_discovery_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Discovery Method', fontsize=12)
plt.ylabel('Log(Number of Planets)', fontsize=12)
plt.title('Logarithmic Number of Planets by Discovery Method', fontsize=14)
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
discovery_counts.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired(np.arange(len(discovery_counts))))
plt.ylabel('')
plt.title('Percentage of Planets by Discovery Method', fontsize=14)

plt.tight_layout()
plt.show()
```
